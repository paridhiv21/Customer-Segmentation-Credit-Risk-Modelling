```python
from sklearn.preprocessing import StandardScaler
df_std = StandardScaler().fit_transform(df)
```

```python
from sklearn.decomposition import PCA
pca = PCA().fit(df_std)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlim(0,16,1)
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
```

```python
from sklearn.decomposition import PCA 
sklearn_pca = PCA(n_components=7)
Y_sklearn = sklearn_pca.fit_transform(df_std)
```

```python
df=pd.DataFrame(Y_sklearn,columns=['PCA1','PCA2','PCA3','PCA4','PCA5','PCA6','PCA7'])
```

```python
from sklearn.cluster import KMeans
wcss = [] 
m = range(1,11)
for i in m: 
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(df) 
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(m,wcss,'bx-')
plt.plot(range(1, 11), wcss)
plt.xlabel('PCA')
plt.ylabel('Inertia') 
plt.show()
```

```python
X = df[["PCA1", "PCA2","PCA3","PCA4","PCA5","PCA6","PCA7"]]
K=3

Centroids = (X.sample(n=K))
```

```python
Centroids
```

```python
diff = 1
j=0

while(diff!=0):
    XD=X
    i=1
    for index1, row_c in Centroids.iterrows():
        ED=[]
        for index2, row_d in XD.iterrows():
            d1 = (row_c["PCA1"]-row_d["PCA1"])**2
            d2 = (row_c["PCA2"]-row_d["PCA2"])**2
            d3 = (row_c["PCA3"]-row_d["PCA3"])**2
            d4 = (row_c["PCA4"]-row_d["PCA4"])**2
            d5 = (row_c["PCA5"]-row_d["PCA5"])**2
            d6 = (row_c["PCA6"]-row_d["PCA6"])**2
            d7 = (row_c["PCA7"]-row_d["PCA7"])**2
         
            d =math.sqrt(d1+d2+d3+d4+d5+d6+d7)
            ED.append(d)
        X[i] = ED
        i = i+1
    
    C = []
    for index, row in X.iterrows():
        min_dist=row[1]
        pos=1
        for i in range(K):
            if row[i+1] < min_dist:
                min_dist = row[i+1]
                pos = i+1
        C.append(pos)
    X["Cluster"]=C
    Centroids_new = X.groupby(["Cluster"]).mean()[["PCA1", "PCA2","PCA3","PCA4","PCA5","PCA6","PCA7"]]
    if j == 0:
        diff = 1
        j = j+1
    else:
        diff = (Centroids_new['PCA1'] - Centroids['PCA1']).sum() + (Centroids_new['PCA2'] - Centroids['PCA2']).sum() + (Centroids_new['PCA3'] - Centroids['PCA3']).sum() + (Centroids_new['PCA4'] - Centroids['PCA4']).sum() +(Centroids_new['PCA5'] - Centroids['PCA5']).sum()+ (Centroids_new['PCA6'] - Centroids['PCA6']).sum()+(Centroids_new['PCA7'] - Centroids['PCA7']).sum()
        print(diff.sum())
    Centroids = X.groupby(["Cluster"]).mean()[["PCA1","PCA2","PCA3","PCA4","PCA5","PCA6","PCA7"]]
```
